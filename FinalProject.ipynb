2. **Predictive Objective: Machine Learning Application**
<br><br>Tujuan dari proyek ini adalah membangun model machine learning untuk membantu bank menentukan apakah seorang pelanggan akan melakukan term deposit, berdasarkan informasi yang tersedia **hingga saat ini**.

    **Model tidak terbatas hanya untuk prediksi kontak pertama**

    Meskipun fitur seperti `campaign` menunjukkan jumlah total kontak yang sudah dilakukan terhadap seorang pelanggan, informasi ini justru merefleksikan **kondisi nyata saat ini** misalnya, "ini adalah kontak ke-2 atau ke-4 untuk pelanggan A." Karena itu:

    - Model **tidak hanya memprediksi pada kontak pertama**
    - Tapi juga memprediksi hasil kontak **saat ini**, terlepas dari sudah ke berapa kali pelanggan dihubungi

    **Justifikasi Pendekatan Ini:**

    - **Tidak ada ID pelanggan** → kita tidak bisa membedakan riwayat per orang
    - **Tidak ada baris eksplisit untuk tiap urutan kontak** → dataset hanya menyimpan snapshot terakhir per pelanggan
    - Maka, **satu baris = satu pelanggan** adalah asumsi paling aman dan praktis
    - Fitur `campaign`, `previous`, dan `poutcome` tetap bisa digunakan karena nilainya sudah tersedia **sebelum prediksi dilakukan**

    **Tujuan kedua adalah membangun model prediktif berbasis machine learning yang mampu:**

    > Memprediksi apakah pelanggan akan melakukan term deposit pada saat mereka dihubungi, dengan mempertimbangkan informasi historis dan status saat ini tanpa menggunakan data masa depan seperti `duration`.

    Model ini akan digunakan untuk:
    - Membantu tim marketing memprioritaskan pelanggan yang paling potensial
    - Mengurangi biaya kampanye dengan menyaring pelanggan yang kemungkinan besar tidak tertarik

Dengan pendekatan ini, bank diharapkan tidak hanya mendapatkan insight retrospektif, tetapi juga kemampuan untuk mengoptimalkan keputusan kampanye secara proaktif ke depannya.

#### **Pendekatan Metriks**

Apabila dilihat dari dataset yang ada terdapat variabel y (target) yang menunjukan sebagai berikut:

Target:

`0 : Nasabah tidak melakukan term deposit`

`1 : Nasabah melakukan term deposit`

`Type 0 Error: Nasabah yang diprediksi tidak melakukan term deposit tetapi pada faktanya nasabah ini melakukan term deposit (False Negative)` -> Konsekuensi: kehilangan nasabah potensial

`Type 1 Error: Nasabah yang diprediksi melakukan term deposit tetapi pada faktanya nasabah ini tidak melakukan term deposit (False Positive)` -> Konsekuensi: Pengeluaran biaya yang tidak tepat sasaran

**Business Metric:**

- Berdasarkan [LiveAgent](https://www.liveagent.com/customer-support-glossary/cost-per-call/), biaya panggilan telepon di portugal adalah 2.34 Euro - 4.85 Euro. Kita asumsikan bahwa cost dari tiap panggilan adalah 3.6 Euro.
- Berdasarkan data dari [Institut Statistik Portugal](https://www.ine.pt/xportal/xmain?PUBLICACOESmodo=2&PUBLICACOESpub_boui=141619793&xpgid=ine_publicacoes&xpid=INE), pada tahun 2010 pendapatan rata rata penduduk per tahunnya adalah 23,811 Euro dengan pengeluaran untuk biaya hidup sebesar 20,391 Euro yang menyisakan disposable income sebesar 3,420 Euro.
- Berdasarkan [Banco De Portugal](https://www.bportugal.pt/sites/default/files/anexos/papers/ab201014_e.pdf), rata rata orang akan menginvestasikan sekitar `2,695 Euro`.
- Berdasarkan [Banco De Portugal](https://www.bportugal.pt/sites/default/files/anexos/pdf-boletim/ref_november10_e.pdf), Bunga pinjaman berada di kisaran 4.2 – 5.1% tergantung dari jangka waktu dan menurut [TradingEconomics](https://tradingeconomics.com/portugal/bank-lending-rate#:~:text=Bank%20Lending%20Rate%20in%20Portugal,source%3A%20European%20Central%20Bank) rata rata lending interest rate di portugal dari tahun 2003-2025 adalah 4.98%. 
- Berdasarkan [TheBanks](https://thebanks.eu/compare-banking-products/time-deposit-accounts/Portugal), deposit interest rate portugal berkisar antara 0.10% - 3% tergantung dari jangka waktunya. Apabila dilihat dari minimal depositonya, kita dapat menyimpulkan bahwa term deposit yang sesuai dengan kriteria yang ada adalah 0.10%

Maka dari itu, dapat disimpulkan bahwa:

Rata-rata investasi:
- `2695 Euro`

Keuntungan per nasabah (Kelas 1):
- `Nominal deposit * (Bunga Pinjaman - Bunga deposit)` -> `2695 * (4.98%-0.10%) = 131 Euro per tahun`

Potensi Keuntungan(berdasarkan jumlah nasabah yang ada pada dataset):
- `131 Euro * 41188 = 5,395,628 Euro` -> apabila semua nasabah melakukan term deposit

Kerugian:
- Type 0 Error -> Apabila kehilangan nasabah yang akan mendepositkan uang sejumlah `2695 Euro`, maka bank akan mendapat kerugian sebesar `131 Euro` dari tiap nasabah yang hilang.
- Type 1 Error -> Apabila bank mengeluarkan biaya untuk nasabah yang tidak melakukan term deposit, maka bank akan mengeluarkan biaya untuk setiap panggilannya. Kita asumsikan jumlah panggilan rata rata per nasabah adalah 5x panggilan berdasarkan data, maka untuk setiap panggilan yang tidak terkonversi, bank akan merugi sekitar `18 Euro`

### Checking Missing Values

Karena missing values ditandai dengan nilai `unknown`, maka kita perlu cek kolom mana saja yang memiliki nilai `unknown`

````python
columns_with_value = []
for col in df.columns:
    if 'unknown' in df[col].values:
        columns_with_value.append(col)
print(f"Columns containing the value 'unknown': {columns_with_value}")
````

````python
missing = []
for cols in columns_with_value:
    missing.append([cols, df[df[cols]=='unknown'].shape[0],(df[df[cols]=='unknown'].shape[0]/41188)*100])
missing_values = pd.DataFrame(columns=['Column Name', 'Number of missing', 'Percentage'],
                     data=missing)
missing_values
````

## **Conversion Rate in Each Category**

### What is Conversion Rate?

Conversion rate adalah persentase pengunjung yang menyelesaikan tindakan yang diinginkan seperti membeli produk, mengisi formulir, dll. Secara sederhana, conversion rate adalah cara untuk mengetahui seberapa efektif upaya pemasaran Anda dalam mengubah pengunjung menjadi pelanggan atau mendorong mereka melakukan tindakan yang Anda inginkan. Conversion rate membantu perusahaan memahami seberapa efektif kampanye pemasaran yang dilakukan dalam mencapai tujuan yang diinginkan. Dengan mengetahui tingkat konversi, perusahaan dapat mengidentifikasi area yang perlu diperbaiki dan mengoptimalkan strategi pemasaran untuk mendapatkan hasil yang lebih baik. 

Rumus Conversion rate:

*Conversion Rate = (Jumlah Konversi / Jumlah Total Pengunjung) * 100%*


Untuk memperjelas conversion rate berdasarkan umur nasabah, kami melakukan binning terhadap data umur yang akan mengelompokan nasabah berdasarkan 6 kategori yaitu:
- 17-24 tahun
- 25-34 tahun
- 35-44 tahun
- 45-54 tahun
- 55-64 tahun
- 65+ tahun

#### Feature Engineering

````python
# Kelompokkan umur ke dalam rentang
bins = [17, 25, 35, 45, 55, 65, 100]
labels = ['17-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
````

### Age Range

Umur sering dikaitkan dengan conversion rate dalam marketing karena umur adalah salah satu indikator utama yang memengaruhi perilaku konsumen, preferensi, dan cara pengambilan keputusan. Umur berpengaruh besar terhadap pola pikir, kebutuhan, daya beli, dan perilaku digital konsumen. Maka, mengaitkan umur dengan conversion rate membantu marketer menyesuaikan strategi agar lebih tepat sasaran dan efektif.


````python
# Hitung total dan yang y = yes
conversion_data = df.groupby('age_group')['y'].value_counts().unstack().fillna(0)
conversion_data['conversion_rate'] = conversion_data['yes'] / (conversion_data['yes'] + conversion_data['no'])
# Sort berdasarkan conversion rate
conversion_data_sorted = conversion_data.sort_values(by='conversion_rate', ascending=False)

# Visualisasi
plt.figure(figsize=(8, 6))
bars = sns.barplot(
    x=conversion_data_sorted.index,
    y=conversion_data_sorted['conversion_rate'],
    hue=conversion_data_sorted.index,
    order=conversion_data_sorted.index
)

# Tambahkan label persentase di atas setiap batang
for bar in bars.patches:
    height = bar.get_height()
    bars.annotate(f'{height:.1%}',
                  xy=(bar.get_x() + bar.get_width() / 2, height),
                  xytext=(0, 5),  # offset di atas bar
                  textcoords='offset points',
                  ha='center', va='bottom', fontsize=10)

# Format plot
plt.title('Conversion Rate berdasarkan Kelompok Usia (Diurutkan)')
plt.xlabel('Kelompok Usia')
plt.ylabel('Conversion Rate')
plt.ylim(0, 1)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

**Penanganan Outlier**

Setelah analisis visual dan statistik terhadap fitur numerik, kita menerapkan transformasi untuk mereduksi pengaruh nilai ekstrim (outlier) terhadap model. Transformasi dilakukan satu per satu dengan justifikasi yang jelas.

Berikut langkah-langkahnya:

**Capping Fitur `campaign`**

- `campaign` merepresentasikan jumlah kontak yang dilakukan kepada nasabah dalam satu kampanye.
- Secara statistik, mayoritas nilai berada di bawah 5, tetapi ada outlier hingga **56 kali**.
- Nilai ekstrem tersebut jarang terjadi dan berpotensi besar menyesatkan model.

Maka kita akan membatasi nilai `campaign` ke batas **P95** (nilai persentil ke-95).

> Ini menghindari distorsi model tanpa menghapus data apa pun.


````python
# Cek distribusi campaign
X['campaign'].describe(percentiles=[0.75, 0.95, 0.96, 0.97, 0.99])
````

````python
# Terapkan capping campaign
import numpy as np

X_capped = X.copy()
campaign_cap = X_capped['campaign'].quantile(0.95)
X_capped['campaign'] = np.where(X_capped['campaign'] > campaign_cap, campaign_cap, X_capped['campaign'])
````

Alasan Capping:

- 75% data = 0 → mayoritas belum pernah merespon campaign
- Namun ada lonjakan ke 7 (maksimum)
- Model bisa terlalu fokus pada minor kasus ekstrem
- P95 cukup konservatif dan membantu hilangkan lonjakan tanpa hilangkan info penting


````python
X['previous'].describe(percentiles=[0.75, 0.95, 0.97, 0.99])
````

````python
previous_cap = X_capped['previous'].quantile(0.95)
X_capped['previous'] = np.where(X_capped['previous'] > previous_cap, previous_cap, X_capped['previous'])
````

Alasan Capping:

- Beberapa nilai mendekati 100 tahun

- Capping di P99 untuk jaga skala

### Alasan Penggunaan Classification Metric dan Business Metric dalam Evaluasi Model

Dalam proyek prediksi term deposit bank, kami menggunakan pendekatan evaluasi bertingkat yang dimulai dengan **classification metric (F2-Score)** dan diakhiri dengan **business metric** sebagai penentu akhir. Dokumen ini menjelaskan alasan filosofis dan praktis di balik pendekatan ini.

### 1. Mengapa Menggunakan F2-Score Sebagai Eliminasi Pertama?

### 1.1 Karakteristik F2-Score
F2-Score memberikan bobot lebih tinggi pada **recall** dibandingkan precision, dengan formula:
```
F2 = (1 + 2²) × (precision × recall) / (2² × precision + recall)
```

### 1.2 Konteks Bisnis Bank
Dalam konteks term deposit bank:
- **False Negative (FN)**: Nasabah yang diprediksi tidak akan berlangganan padahal sebenarnya akan berlangganan
- **False Positive (FP)**: Nasabah yang diprediksi akan berlangganan padahal sebenarnya tidak akan berlangganan

**Mengapa FN lebih kritis?**
- Kehilangan nasabah potensial yang sebenarnya tertarik
- Opportunity cost yang tinggi
- Dampak jangka panjang terhadap revenue

### 1.3 F2-Score sebagai Filter Awal
F2-Score digunakan sebagai **filter teknis** untuk:
- Menghilangkan model dengan performa recall yang buruk
- Memastikan model dapat mengidentifikasi sebagian besar nasabah potensial
- Memberikan baseline minimum untuk evaluasi lanjutan

### Definisikan Business Metric:

````python
from sklearn.metrics import confusion_matrix, make_scorer

def business_metric(y, y_pred):
    tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()
    total = tn + fp + fn + tp
    raw_score = ((tp*131 - (fp+tp)*18) - ((tp+fn)*131 - total*18)) / total
    # Normalisasi skor ke 0–1
    min_score = -14.52
    max_score = 18.0
    normalized_score = (raw_score - min_score) / (max_score - min_score)
    return normalized_score

business_score = make_scorer(business_metric, greater_is_better=True)
````

## **Threshold Optimization**

````python
# ==========================================
# THRESHOLD OPTIMIZATION EXPERIMENT
# ==========================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    fbeta_score, recall_score, precision_score, 
    confusion_matrix, classification_report,
    precision_recall_curve, roc_curve, auc
)

# Business context dari notebook Anda
REVENUE_PER_CUSTOMER = 131  # Euro (profit dari term deposit)
COST_PER_CALL = 18          # Euro (biaya telepon)

# 1. Dapatkan probability scores dari best model
print("🔍 Getting probability scores from best model...")
y_proba = lgbm_search.best_estimator_.predict_proba(X_test)[:, 1]

# 2. Define threshold range untuk eksperimen
thresholds = np.arange(0.1, 1.0, 0.05)  # 0.1, 0.15, 0.2, ..., 0.95
print(f"📊 Testing {len(thresholds)} thresholds: {thresholds}")

# 3. Eksperimen threshold
print("\n⚡ Starting threshold optimization experiment...")
threshold_results = []

for threshold in thresholds:
    # Prediksi dengan threshold tertentu
    y_pred = (y_proba >= threshold).astype(int)
    
    # Hitung metrics
    f2_score = fbeta_score(y_test, y_pred, beta=2)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    
    # Hitung confusion matrix components
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    
    # Business cost calculation (YANG BERUBAH)
    fn_cost = fn * REVENUE_PER_CUSTOMER  # Kehilangan revenue
    fp_cost = fp * COST_PER_CALL         # Waste cost
    total_cost = fn_cost + fp_cost
    
    # Net profit calculation (TAMBAHAN)
    revenue = tp * REVENUE_PER_CUSTOMER
    total_call_cost = (tp + fp) * COST_PER_CALL
    net_profit = revenue - total_call_cost
    
    threshold_results.append({
        'threshold': threshold,
        'f2_score': f2_score,
        'recall': recall,
        'precision': precision,
        'tp': tp,
        'tn': tn,
        'fp': fp,
        'fn': fn,
        'fn_cost': fn_cost,
        'fp_cost': fp_cost,
        'total_cost': total_cost,
        'revenue': revenue,
        'total_call_cost': total_call_cost,
        'net_profit': net_profit
    })

# 4. Convert ke DataFrame untuk analisis
threshold_df = pd.DataFrame(threshold_results)
print(f"\n✅ Experiment completed! Results for {len(threshold_df)} thresholds")

# 5. Analisis hasil
print("\n" + "="*60)
print("📈 THRESHOLD OPTIMIZATION RESULTS")
print("="*60)

# Find optimal threshold berdasarkan F2-Score
optimal_f2_idx = threshold_df['f2_score'].idxmax()
optimal_f2_threshold = threshold_df.loc[optimal_f2_idx, 'threshold']
optimal_f2_score = threshold_df.loc[optimal_f2_idx, 'f2_score']

print(f"\n🎯 OPTIMAL THRESHOLD (F2-Score): {optimal_f2_threshold:.3f}")
print(f"   F2-Score: {optimal_f2_score:.4f}")
print(f"   Recall: {threshold_df.loc[optimal_f2_idx, 'recall']:.4f}")
print(f"   Precision: {threshold_df.loc[optimal_f2_idx, 'precision']:.4f}")
print(f"   Net Profit: {threshold_df.loc[optimal_f2_idx, 'net_profit']:.0f} Euro")

# Find optimal threshold berdasarkan Net Profit
optimal_profit_idx = threshold_df['net_profit'].idxmax()
optimal_profit_threshold = threshold_df.loc[optimal_profit_idx, 'threshold']
optimal_profit = threshold_df.loc[optimal_profit_idx, 'net_profit']

print(f"\n💰 OPTIMAL THRESHOLD (Net Profit): {optimal_profit_threshold:.3f}")
print(f"   Net Profit: {optimal_profit:.0f} Euro")
print(f"   F2-Score: {threshold_df.loc[optimal_profit_idx, 'f2_score']:.4f}")

# 6. Visualisasi hasil
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Threshold Optimization Analysis', fontsize=16, fontweight='bold')

# Plot 1: F2-Score vs Threshold
axes[0, 0].plot(threshold_df['threshold'], threshold_df['f2_score'], 'b-', linewidth=2)
axes[0, 0].axvline(x=optimal_f2_threshold, color='red', linestyle='--', 
                   label=f'Optimal F2: {optimal_f2_threshold:.3f}')
axes[0, 0].set_xlabel('Threshold')
axes[0, 0].set_ylabel('F2-Score')
axes[0, 0].set_title('F2-Score vs Threshold')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].legend()

# Plot 2: Net Profit vs Threshold
axes[0, 1].plot(threshold_df['threshold'], threshold_df['net_profit'], 'g-', linewidth=2)
axes[0, 1].axvline(x=optimal_profit_threshold, color='red', linestyle='--', 
                   label=f'Optimal Profit: {optimal_profit_threshold:.3f}')
axes[0, 1].set_xlabel('Threshold')
axes[0, 1].set_ylabel('Net Profit (Euro)')
axes[0, 1].set_title('Net Profit vs Threshold')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].legend()

# Plot 3: Cost Analysis
axes[1, 0].plot(threshold_df['threshold'], threshold_df['fn_cost'], 'red', 
                linewidth=2, label='FN Cost (Revenue Loss)')
axes[1, 0].plot(threshold_df['threshold'], threshold_df['fp_cost'], 'blue', 
                linewidth=2, label='FP Cost (Waste)')
axes[1, 0].set_xlabel('Threshold')
axes[1, 0].set_ylabel('Cost (Euro)')
axes[1, 0].set_title('Cost Analysis vs Threshold')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].legend()

# Plot 4: Precision-Recall vs Threshold
axes[1, 1].plot(threshold_df['threshold'], threshold_df['precision'], 'g-', 
                linewidth=2, label='Precision')
axes[1, 1].plot(threshold_df['threshold'], threshold_df['recall'], 'orange', 
                linewidth=2, label='Recall')
axes[1, 1].axvline(x=optimal_f2_threshold, color='red', linestyle='--')
axes[1, 1].set_xlabel('Threshold')
axes[1, 1].set_ylabel('Score')
axes[1, 1].set_title('Precision & Recall vs Threshold')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].legend()

plt.tight_layout()
plt.show()

# 7. Detailed comparison table
print("\n�� DETAILED THRESHOLD COMPARISON")
print("="*60)
comparison_df = threshold_df[
    (threshold_df['threshold'] == 0.5) |  # Default threshold
    (threshold_df['threshold'] == optimal_f2_threshold) |  # Optimal F2
    (threshold_df['threshold'] == optimal_profit_threshold)  # Optimal Profit
].round(4)

comparison_df['threshold'] = comparison_df['threshold'].map({
    0.5: 'Default (0.5)',
    optimal_f2_threshold: f'Optimal F2 ({optimal_f2_threshold:.3f})',
    optimal_profit_threshold: f'Optimal Profit ({optimal_profit_threshold:.3f})'
})

print(comparison_df[['threshold', 'f2_score', 'recall', 'precision', 'net_profit']].to_string(index=False))

# 8. Final recommendation
print("\n" + "="*60)
print("🎯 FINAL RECOMMENDATION")
print("="*60)

if optimal_f2_threshold == optimal_profit_threshold:
    print(f"✅ RECOMMENDED THRESHOLD: {optimal_f2_threshold:.3f}")
    print("   (Same for both F2-Score and Net Profit optimization)")
else:
    print(f"🎯 For F2-Score optimization: {optimal_f2_threshold:.3f}")
    print(f"💰 For Net Profit optimization: {optimal_profit_threshold:.3f}")
    print(f"\n💡 RECOMMENDATION: Use {optimal_f2_threshold:.3f} (F2-Score optimal)")
    print("   Reason: Prioritizes business goal of minimizing False Negatives")

# 9. Apply optimal threshold untuk final prediction
print(f"\n🚀 Applying optimal threshold {optimal_f2_threshold:.3f} to final model...")
final_y_pred = (y_proba >= optimal_f2_threshold).astype(int)

print("\n📋 FINAL MODEL PERFORMANCE WITH OPTIMAL THRESHOLD:")
print("="*60)
print(classification_report(y_test, final_y_pred, target_names=['No', 'Yes']))

# 10. Save optimal threshold untuk future use
optimal_threshold = optimal_f2_threshold
print(f"\n💾 Optimal threshold {optimal_threshold:.3f} saved for future predictions")
````

### **Evaluasi Model dan Teknik Resampling**

**Random Over-Sampling (ROS)** berhasil meningkatkan representasi kelas minoritas dengan menduplikasi sampel positif, namun berpotensi menyebabkan overfitting. **Random Under-Sampling (RUS)** efektif dalam menyeimbangkan dataset namun berpotensi kehilangan informasi penting dari kelas mayoritas. **SMOTE (Synthetic Minority Over-sampling Technique)** menghasilkan sampel sintetik yang realistic namun dalam beberapa kasus dapat menyebabkan noise dalam data.

Setelah evaluasi menyeluruh terhadap semua kombinasi model dan teknik resampling, tiga kombinasi terbaik yang berhasil mengatasi ketidakseimbangan kelas dengan optimal adalah:

1. **ROS + LGBM** - Menunjukkan performa superior dengan F2-score tertinggi (0.57)
2. **ROS + Gradient Boost** - Memberikan hasil yang solid dengan recall yang tinggi (0.66) dan kemampuan menangani overfitting
3. **RUS + Gradient Boost** - Menunjukkan performa yang baik namun mengalami sedikit penurunan dalam F2

Untuk menentukan model terbaik yang akan digunakan dalam implementasi bisnis, kami melakukan eliminasi lanjutan menggunakan business metrics yang mencakup net profit, ROI improvement, dan cost-benefit analysis. **ROS + LGBM** akhirnya dipilih sebagai best model karena berhasil mengoptimalkan trade-off antara metrik teknis dan dampak bisnis, dengan hasil test business score tertinggi yaitu 0.739.

Model ini kemudian menjalani proses hyperparameter tuning yang intensif untuk memastikan performa optimal dalam lingkungan produksi. Tuning ini mencakup optimasi learning rate, number of estimators, max depth, dan parameter regularization untuk memastikan model yang robust dan scalable.

Salah satu aspek yang paling menantang dalam proyek ini adalah penentuan threshold optimal. Kami menyadari bahwa menggunakan threshold default 0.5 belum tentu atau setidak belum pernah kita buktikan tidak optimal untuk kasus ini karena ketidakseimbangan kelas. Melalui analisis yang mendalam terhadap berbagai metrik evaluasi, kami melakukan testing terhadap range threshold dari 0.1 hingga 0.9 dengan interval 0.05.

Evaluasi threshold dilakukan dengan mempertimbangkan berbagai metrik termasuk precision, recall, F1-score, F2-score, dan business metrics seperti net profit. **F2-score** dipilih sebagai metrik utama karena memberikan bobot lebih tinggi pada recall, yang kritis dalam konteks bisnis dimana false negative (kehilangan nasabah potensial) lebih mahal daripada false positive.

Setelah analisis komprehensif, threshold **0.500** dipilih sebagai optimal karena memberikan keseimbangan terbaik antara F2-score (0.58) dan net profit (52.643 Euro). Threshold ini berhasil mengoptimalkan business impact sambil mempertahankan performa model yang solid.
