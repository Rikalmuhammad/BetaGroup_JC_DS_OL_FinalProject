```python
# Cek distribusi menggunakan Shapiro Wilk Test
from scipy.stats import shapiro

for cols in num_cols:
    stat, p = shapiro(df[cols])
    print('Kolom:', cols)
    print('Statistik:', stat)
    print('p-value:', p)
    
    if p > 0.05:
        print("Data terdistribusi normal (gagal tolak H0)")
    else:
        print("Data tidak terdistribusi normal (tolak H0)")
    print('\n')
```

## **Numerical Feature Correlation**

```python
import numpy as np
# Pilih fitur numerik dan ubah target menjadi numerik
numerical_df = df.select_dtypes(include=['int64', 'float64'])
numerical_df['y'] = df['y'].map({'no': 0, 'yes': 1})

# Hitung korelasi
corr = numerical_df.corr()

# Buat mask untuk segitiga atas
mask = np.triu(np.ones_like(corr, dtype=bool))

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, mask=mask, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix (Lower Triangle Only)', fontsize=14)
plt.tight_layout()
plt.show()
```

Insights:
- Korelasi kuat antar indikator ekonomi(euribor3m, emp.var.rate, dan nr.employed). Hal ini menunjukkan bahwa indikator ekonomi tersebut sangat terhubung dan cenderung bergerak bersama.

## **Conversion Rate in Each Category**

### What is Conversion Rate?

Conversion rate adalah persentase pengunjung yang menyelesaikan tindakan yang diinginkan seperti membeli produk, mengisi formulir, dll. Secara sederhana, conversion rate adalah cara untuk mengetahui seberapa efektif upaya pemasaran Anda dalam mengubah pengunjung menjadi pelanggan atau mendorong mereka melakukan tindakan yang Anda inginkan. Conversion rate membantu perusahaan memahami seberapa efektif kampanye pemasaran yang dilakukan dalam mencapai tujuan yang diinginkan. Dengan mengetahui tingkat konversi, perusahaan dapat mengidentifikasi area yang perlu diperbaiki dan mengoptimalkan strategi pemasaran untuk mendapatkan hasil yang lebih baik. 

Rumus Conversion rate:

*Conversion Rate = (Jumlah Konversi / Jumlah Total Pengunjung) * 100%*


Untuk memperjelas conversion rate berdasarkan umur nasabah, kami melakukan binning terhadap data umur yang akan mengelompokan nasabah berdasarkan 6 kategori yaitu:
- 17-24 tahun
- 25-34 tahun
- 35-44 tahun
- 45-54 tahun
- 55-64 tahun
- 65+ tahun

#### Feature Engineering

```python
# Kelompokkan umur ke dalam rentang
bins = [17, 25, 35, 45, 55, 65, 100]
labels = ['17-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

### Age Range

Umur sering dikaitkan dengan conversion rate dalam marketing karena umur adalah salah satu indikator utama yang memengaruhi perilaku konsumen, preferensi, dan cara pengambilan keputusan. Umur berpengaruh besar terhadap pola pikir, kebutuhan, daya beli, dan perilaku digital konsumen. Maka, mengaitkan umur dengan conversion rate membantu marketer menyesuaikan strategi agar lebih tepat sasaran dan efektif.


```python
# Hitung total dan yang y = yes
conversion_data = df.groupby('age_group')['y'].value_counts().unstack().fillna(0)
conversion_data['conversion_rate'] = conversion_data['yes'] / (conversion_data['yes'] + conversion_data['no'])
# Sort berdasarkan conversion rate
conversion_data_sorted = conversion_data.sort_values(by='conversion_rate', ascending=False)

# Visualisasi
plt.figure(figsize=(8, 6))
bars = sns.barplot(
    x=conversion_data_sorted.index,
    y=conversion_data_sorted['conversion_rate'],
    hue=conversion_data_sorted.index,
    order=conversion_data_sorted.index
)

# Tambahkan label persentase di atas setiap batang
for bar in bars.patches:
    height = bar.get_height()
    bars.annotate(f'{height:.1%}',
                  xy=(bar.get_x() + bar.get_width() / 2, height),
                  xytext=(0, 5),  # offset di atas bar
                  textcoords='offset points',
                  ha='center', va='bottom', fontsize=10)

# Format plot
plt.title('Conversion Rate berdasarkan Kelompok Usia (Diurutkan)')
plt.xlabel('Kelompok Usia')
plt.ylabel('Conversion Rate')
plt.ylim(0, 1)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Insights:

- `Kelompok usia 65+ memiliki conversion rate tertinggi (47.2%)`. Hal ini terjadi karena mereka cenderung menghindari risiko dan lebih memilih produk keuangan yang aman seperti term deposit dibandingkan produk berisiko tinggi seperti saham atau reksa dana. Ini menunjukkan bahwa segmen lansia adalah target paling responsif dan potensial untuk produk term deposit. Namun, kelompok umur ini tidak bisa dijadikan segmen paling potensial secara bisnis karena banyak dari mereka sudah di masa pasca-produktif sehingga mereka cenderung menyimpan dana yang sudah ada, bukan menambahnya secara rutin.
- `Kelompok usia 17–24 memiliki konversi cukup tinggi (23.97%)`.
- `Kelompok usia produktif (25–44) mendominasi kampanye, tapi memberikan konversi rendah (12.2%)`. Karena kelompok ini berada di usia produktif, mereka memiliki kemungkinan untuk memiliki ketertarikan pada produk investasi dengan return lebih tinggi serta lebih kritis dan selektif terhadap produk keuangan yang menjadikan kedua hal ini indikasi bahwa perlu adanya evaluasi pendekatan terhadap kelompok usia produktif


### Job

Conversion rate dalam telemarketing bank sangat dipengaruhi oleh jenis pekerjaan nasabah. Hal ini karena jenis pekerjaan mencerminkan kestabilan finansial, preferensi risiko, dan kebutuhan likuiditas, yang semuanya berpengaruh langsung terhadap minat dan kemampuan nasabah untuk membeli produk term deposit. 

Term deposit mengharuskan dana dikunci selama periode tertentu. Oleh karena itu, nasabah yang memiliki penghasilan tetap dan stabil (seperti pegawai swasta) lebih percaya diri untuk menyisihkan dana dalam jangka waktu tertentu tanpa mengganggu kebutuhan sehari-hari. Sebaliknya, pekerja lepas, freelancer, atau pengusaha yang menghadapi pendapatan fluktuatif akan lebih memilih produk keuangan yang likuid (seperti tabungan atau reksa dana pasar uang).


```python
df[df['y']=='yes']['campaign']
```

```python
# Hitung jumlah total dan jumlah yang 'yes' untuk tiap pekerjaan
job_conversion = df.groupby('job')['y'].value_counts().unstack().fillna(0)
job_conversion['conversion_rate'] = job_conversion['yes'] / (job_conversion['yes'] + job_conversion['no'])

# Urutkan dari yang tertinggi
job_conversion_sorted = job_conversion.sort_values(by='conversion_rate', ascending=False)

# Visualisasi
plt.figure(figsize=(10, 6))
bars = sns.barplot(
    x=job_conversion_sorted.index,
    y=job_conversion_sorted['conversion_rate'],
    palette='flare'
)

# Tambahkan label persentase di atas batang
for bar in bars.patches:
    height = bar.get_height()
    bars.annotate(f'{height:.1%}',
                  xy=(bar.get_x() + bar.get_width() / 2, height),
                  xytext=(0, 5),
                  textcoords='offset points',
                  ha='center', va='bottom', fontsize=10)

# Format plot
plt.title('Conversion Rate berdasarkan Pekerjaan (job)')
plt.xlabel('Pekerjaan')
plt.ylabel('Conversion Rate')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Insights:
- Mahasiswa dan pensiunan memiliki conversion rate paling tinggi, hal ini terjadi karena beberapa hal ini:
    - Mahasiswa belum memiliki pengeluaran tetap yang besar
    - Mahasiswa memiliki kecenderungan untuk memikirkan/membangun masa depannya 
    - Mahasiswa dan Pensiunan memiliki sumber penghasilan yang terbatas sehingga melihat term deposit ini sebagai sarana investasi yang rendah resiko
    - Mahasiswa dan Pensiunan cenderung lebih berhati-hati dalam mengalirkan uangnya dan cenderung memilih investasi yang memiliki resiko yang rendah
    
    *Apabila dilihat dari jumlah nasabah yang dihubungi dengan nasabah yang melakukan term deposit, mahasiswa dan pensiunan memiliki jumlah yang jauh lebih kecil dibandingkan dengan pekerjaan lain seperti (admin, management, services, dll.) hal ini berkaitan dengan teori pendapatan dan konsumsi nya Keynes yang menjelaskan bahwa ketika pendapatan naik akan ada 2 hal yang ikut naik juga yaitu consumption dan saving, tetapi proporsi konsumsi dan saving itu sendiri akan sangat dipengaruhi oleh tingkat kenaikan pendapatan itu sendiri. ketika pendapatan seseorang masih relatif rendah, ia akan cenderung lebih banyak untuk meningkatkan konsumsinya,dimana hal ini sejalan dengan teori hierarki Maslowe. Orang-orang dengan kenaikan pendapatan rendah akan cenderung berusaha untuk memenuhi kebutuhannya(safety: terhindar dari rasa lapar, dll.), hal ini akhirnya menyebabkan mereka akan lebih berhati-hati ketika menyimpan uang/investasi. Maka dari itu orang-orang seperti mahasiswa dan pensiunan akan cenderung untuk memanfaatkan instrumen investasi yang memiliki resiko rendah.*
- Calon nasabah yang belum memiliki pekerjaan memiliki conversion rate yang lebih tinggi dibanding dengan calon nasabah yang sudah memiliki pekerjaan, hal ini dapat terjadi karena calon nasabah ini sementara ini belum memiliki pekerjaan tetapi memiliki uang dan ingin mengamankan uang yang mereka miliki
- Calon nasabah yang memiliki pekerjaan dilingkup yang lebih professional memiliki conversion rate yang lebih rendah dibanding mahasiswa, pensiunan dan pengangguran, hal ini dapat terjadi karena calon nasabah yang memiliki pekerjaan ini cenderung lebih kritis dengan penawaran bank dan mungkin juga sudah memiliki produk keuangan lainnya.
